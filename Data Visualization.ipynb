```python
data_sma_groupby2 = data_drop_bandung2.groupby(by='kemendagri_nama_kecamatan').sum()
data_sma_groupby2['jumlah_guru_sma']
```

```python
data_merge = pd.merge(data_sma_groupby,data_sma_groupby2,
                      left_on='kemendagri_nama_kecamatan',
                      right_on='kemendagri_nama_kecamatan')
data_merge
```

jika kita melakukan integrasi data dengan menggabungkan 2 data diatas maka kita dapat melihat bawa jumlah sekolah yang terdapat di kecamatan kabupaten bandung saling terhubung dengan jumlah guru yang terdapat di kecamatan kabupaten bandung.

```python
data_kab_bandung = list(data_merge.index)
y1=data_merge['jumlah_sma']
y2=data_merge['jumlah_guru_sma']

x = np.arange(len(data_merge.index))
width = 0.35

fig, ax = plt.subplots(figsize=(32, 7))

ax.set_title('PERBANDINGAN JUMLAH GURU DAN JUMLAH SEKOLAH DI KAB BANDUNG', size=16)
plt.plot(x,y1,label='line jumlah sma')
plt.plot(x,y2,label='line jumlah guru')

plt.plot()

plt.legend()
plt.show()
```

```python
data_kab_bandung = list(data_merge.index)
sma = data_merge['jumlah_sma']
guru = data_merge['jumlah_guru_sma']

x = np.arange(len(data_merge.index))
width = 0.35

fig, ax = plt.subplots(figsize=(62, 7))

SMA = ax.bar(x - width/2, sma, width, label='JUMLAH_SMA', color='steelblue')
guru = ax.bar(x + width/2, guru, width, label='JUMLAH_GURU_SMA', color='lightcoral')

ax.set_title('PERBANDIGAN JUMLAH GURU DAN JUMLAH SEKOLAH DI KAB BANDUNG', size=16)
ax.set_ylabel('JUMLAH SEKOLAH & JUMLAH GURU', size=14)
ax.set_xticks(x)
ax.set_xticklabels(data_kab_bandung, size=12)
ax.legend(fontsize=14)

plt.show()
```
